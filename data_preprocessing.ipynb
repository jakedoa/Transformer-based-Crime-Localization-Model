```python
import pandas as pd
pd.set_option('display.max_columns', None)


df = pd.read_csv('dataset/Chicago_Crimes.csv')
df
```

```python
df.columns
```

```python
# Droping location attributes

df.drop(columns=['X Coordinate', 'Y Coordinate', 'Latitude', 
                 'Longitude', 'Location', 
                 'Updated On', 'Date', 'Year', 
                 'ID', 'Case Number', 'Block',], inplace=True)
df
```

```python
print(len(df[df['District'].isnull()]))
```

```python
# Dropping rows with NaNs

df.dropna(inplace=True)
df.isna().sum()
```
